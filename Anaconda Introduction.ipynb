```python
import matplotlib.pyplot as plt
import pandas as pd

def fibonacci(n):
    fib_seq = [0, 1]  # Initialize the Fibonacci sequence with the first two numbers
    for i in range(2, n):
        fib_seq.append(fib_seq[i-1] + fib_seq[i-2])  # Generate the next Fibonacci number
    return fib_seq

fib_numbers = fibonacci(100)

plt.plot(fib_numbers[:100])  # Plot the first 100 Fibonacci numbers
plt.xlabel('Index')
plt.ylabel('Fibonacci Number')
plt.title('First 100 Fibonacci Numbers')
plt.show()
```
